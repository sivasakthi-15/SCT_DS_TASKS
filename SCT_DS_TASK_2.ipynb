```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/MyDrive/SCT TASKS/Titanic.csv')
data
```

```python
print("Dataset Overview:")
print(data.info())
```

```python
print("\nFirst few rows:")
print(data.head())
```

```python
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
sns.countplot(x='Survived', data=data, palette='coolwarm')
plt.title('Survival Count')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=data, ci=None, palette='coolwarm')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
sns.histplot(data['Age'].dropna(), bins=30, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=data, palette='coolwarm')
plt.title('Age vs Survival')
plt.show()
```

```python
sns.pairplot(data, hue="Survived", diag_kind="kde", palette="Dark2")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=data, alpha=0.6, palette="coolwarm")
plt.title("Scatter Plot of Age vs Fare by Survival Status")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.yscale("log")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
age_survival = data.groupby("Age", as_index=False)["Survived"].mean()
sns.lineplot(x="Age", y="Survived", data=age_survival, color="blue")
plt.title("Survival Rate over Age")
plt.xlabel("Age")
plt.ylabel("Survival Rate")
plt.show()
```
