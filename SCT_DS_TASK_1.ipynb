```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
```

```python
# Load the dataset
df=pd.read_csv("/content/drive/MyDrive/SCT TASKS/world_population.csv")
df
```

```python
print("Dataset Statistics")
df.head()
```

```python
print(" Dataset Information: ")
df.info()
```

```python
print("Summary Statistics: " )
df.describe()
```

```python
print("Checking for Missing Values: ")
df.isnull().sum()
```

```python
# Bar Chart: Population by Continent
continent_population = df.groupby("Continent")["2022 Population"].sum()

plt.figure(figsize=(10, 6))
plt.bar(continent_population.index, continent_population.values, color=['gold', 'royalblue', 'crimson', 'green', 'purple'])
plt.xlabel("Continent", fontsize=12, fontweight='bold')
plt.ylabel("Population (Billions)", fontsize=12, fontweight='bold')
plt.title("🌍 Population Distribution by Continent", fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Download Natural Earth dataset (one-time download)
world = gpd.read_file("https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip")

# Merge with your population dataset
world = world.merge(df, how="left", left_on="ADMIN", right_on="Country/Territory")

# Plot the world map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.plot(column='2022 Population', cmap='coolwarm', linewidth=0.8, edgecolor='black', legend=True, ax=ax)
plt.title("🗺 World Population Distribution (2022)", fontsize=14, fontweight='bold')
plt.show()
```

```python
# Histogram: Population Density Distribution
plt.figure(figsize=(10, 6))
plt.hist(df["Density (per km²)"].dropna(), bins=30, color='teal', edgecolor='black', alpha=0.7)
plt.xlabel("Population Density (per km²)", fontsize=12, fontweight='bold')
plt.ylabel("Number of Countries", fontsize=12, fontweight='bold')
plt.title("📊 Population Density Distribution", fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
